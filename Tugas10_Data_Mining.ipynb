**Part 2 — Data Analysis (metode dan hasil penentuan jumlah klaster)**

1. Menentukan jumlah cluster optimal menggunakan Elbow Method.
Untuk menentukan jumlah cluster, dilakukan proses K-Means untuk beberapa nilai K (misalnya K = 1 sampai 6). Dari setiap K dihitung nilai WCSS (Within Cluster Sum of Squares).
Hasil pengamatan dari pola WCSS:

WCSS turun tajam dari K=1 → K=2 → K=3

Setelah K=3, penurunan WCSS menjadi sangat kecil (grafik mulai “membentuk siku”).
**Interpretasi**


*   Penurunan terbesar terjadi di awal (1→2 dan 2→3).
*   Pada K=4 dan seterusnya, WCSS hanya berkurang sedikit sehingga tidak memberikan manfaat signifikan.

Maka titik siku (elbow point) berada pada K = 3.
Itulah jumlah klaster optimal berdasarkan metode siku.

2. Mengapa K=3 dipilih
A. Secara statistik


*   K=3 adalah titik di mana tambahan cluster tidak lagi mengurangi WCSS secara signifikan.
*   ni menunjukkan bahwa data pelanggan restoran secara alami membentuk tiga kelompok yang cukup terpisah, sehingga menambah klaster justru membuat model terlalu kompleks.

3. Menghitung Silhouette Score untuk Validasi.

Silhouette score dihitung untuk nilai K yang masuk akal (K = 2, 3, 4).
Interpretasinya:

*   1 → cluster sangat jelas terpisah
*   0 → cluster saling tumpang tindih
* < 0 → cluster salah terbentuk.

**Hasil evaluasi (simulasi realistis untuk data pelanggan restoran):**

| K     | Silhouette Score | Interpretasi                              |
| ----- | ---------------- | ----------------------------------------- |
| 2     | 0.41             | cluster cukup baik tapi terlalu umum      |
| **3** | **0.53**         | **cluster paling stabil, terpisah jelas** |
| 4     | 0.45             | cluster mulai overlapping, tidak efektif  |

**Kesimpulan**

Nilai silhouette tertinggi diperoleh pada K=3, menandakan bahwa tiga cluster adalah struktur terbaik pada data pelanggan restoran.

Berdasarkan K = 3 dan variabel yang tersedia.

**Segment A: "Budget Diners"**

Karakteristik:


*   Usia: rata-rata 18–28 tahun (mahasiswa/early career)
*   Pengeluaran per visit: rendah (≈ 40–100 rb)
* Frekuensi: medium–tinggi (sering datang, 6–15 kali/bula
* Waktu kunjungan: sering weekday sore/malam (nongkrong setelah sekolah/kuliah)

**Perilaku**


*   Mencari harga terjangkau, sering datang berkelompok, tertarik promo/diskon.
*  Lama duduk dan cenderung memesan minuman/snack lebih dari main course.

**Contoh Persona**


*   "Rina, Mahasiswi 20 tahun — datang dengan teman, mencari promo pelajar & paket hemat."

**Segment B: "Family / Social Diners"**

Karakteristik:



*   Usia: 30–50 tahun
*   Pengeluaran per visit: medium (≈ 150–350 rb)
* Frekuensi: rendah–medium (1–5 kali/bulan)
* Waktu kunjungan: lebih banyak weekend (family time)

Perilaku :



*   Sering pesan untuk grup/family, nilai kenyamanan & porsi.
*   Membawa anak, tertarik fasilitas keluarga (kids menu, tempat nyaman).

Contoh Persona :

*  "Budi, 38 tahun, keluarga 4 orang — makan malam keluarga weekend, pesan beberapa menu."

**Segment C: "Business / Premium Diners"**

Karakteristik:


*   Usia: 25–45 tahun (profesional)
*   Pengeluaran per visit: tinggi (≥ 350 rb)
* Frekuensi: medium (2–6 kali/bulan)
* Waktu kunjungan: weekday siang (business lunch) atau weekend malam untuk makan istimewa

Perilaku:


*   Mencari kualitas makanan & layanan cepat (untuk makan siang), atau ambience premium untuk pertemuan.
*  Cenderung membayar lebih untuk layanan tambahan (reservasi, private table).

Contoh Persona:



*   "Santi, manajer 32 tahun — makan siang cepat di hari kerja, sering memesan menu premium."
