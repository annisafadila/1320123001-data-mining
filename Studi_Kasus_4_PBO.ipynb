Studi Kasus 4: Sistem Penjualan Atribut: namaProduk, harga, stok Method: tampilkanProduk() Buat minimal 5 object produk


Pada studi kasus ini kita membuat class Produk untuk menyimpan data barang pada sistem penjualan.

Atribut:
namaProduk
harga
stok
Method:
tampilkanProduk() → menampilkan data produk
Ketentuan:

Membuat minimal 5 object produk.

```python
public class Produk {

    // Atribut
    String namaProduk;
    int harga;
    int stok;

    // Constructor
    Produk(String namaProduk, int harga, int stok) {
        this.namaProduk = namaProduk;
        this.harga = harga;
        this.stok = stok;
    }

    // Method tampilkan produk
    void tampilkanProduk() {
        System.out.println("Nama Produk : " + namaProduk);
        System.out.println("Harga       : Rp " + harga);
        System.out.println("Stok        : " + stok);
        System.out.println("------------------------");
    }

    public static void main(String[] args) {

        // Membuat 5 object produk
        Produk p1 = new Produk("Laptop", 7500000, 10);
        Produk p2 = new Produk("Mouse", 150000, 25);
        Produk p3 = new Produk("Keyboard", 300000, 15);
        Produk p4 = new Produk("Printer", 2200000, 8);
        Produk p5 = new Produk("Flashdisk", 90000, 30);

        // Menampilkan data produk
        p1.tampilkanProduk();
        p2.tampilkanProduk();
        p3.tampilkanProduk();
        p4.tampilkanProduk();
        p5.tampilkanProduk();
    }
}
```

🎯 Kesimpulan
Class = Produk
Object = p1, p2, p3, p4, p5
Atribut = namaProduk, harga, stok
Method = tampilkanProduk()
