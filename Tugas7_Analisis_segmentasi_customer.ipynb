**KASUS B -SEGMENTASI MAHASISWA BERDASARKAN PERILAKU BELAJAR**

**BAGIAN 1: IDENTIFIKASI POLA**

POLA 1: Mahasiswa Rajin dan Berprestasi.

*   **Karakteristik**: IPK tinggi (≥3.5), tingkat kehadiran di atas 85%, jam belajar di atas 15 jam/minggu, partisipasi kelas aktif, dan sering mengakses e-learning.
*   **Contoh** : Mahasiswa dengan NIM 001 (IPK 3.8, kehadiran 95%, jam belajar 20 jam/minggu).

POLA 2: Mahasiswa Cukup Aktif dengan Konsistensi Sedang.

*   **Karakteristik** : IPK 3.0–3.5, kehadiran 70–85%, jam belajar 10–15 jam/minggu, partisipasi kelas sedang, dan akses e-learning tidak rutin.
*   **Contoh** : Mahasiswa dengan NIM 003 (IPK 3.5, kehadiran 85%, jam belajar 15 jam/minggu).

POLA 3: Mahasiswa Kurang Aktif dan Perlu Dukungan Tambahan.


*   **Karakteristik** : IPK kurang dari 3.0, kehadiran rendah ,kurang dari 75%, jam belajar di bawah 10 jam/minggu, partisipasi kelas rendah, dan jarang menggunakan e-learning.
*   **Contoh** : Mahasiswa dengan NIM 002 (IPK 2.9, kehadiran 70%, jam belajar 10 jam/minggu).









**BAGIAN 2: SEGMENTASI MAHASISWA**

SEGMENT 1: Mahasiswa Unggul (High Achievers)

**Jumlah:** ±30% dari total mahasiswa

**Ciri Khas**: IPK ≥ 3.5 , Disiplin dan aktif mengikuti perkuliahan , Sering berpartisipasi dalam diskusi dan tugas kelompok , Konsisten mengakses e-learning dan bahan tambahan.

**Contoh Representative**: NIM 001

SEGMENT 2: Mahasiswa Konsisten (Moderate Performers)

**Jumlah**: ±40% dari total mahasiswa

**Ciri Khas**: IPK 3.0–3.5, Kehadiran cukup baik (75–85%) ,Partisipasi kelas sedang, Belajar terjadwal namun belum maksimal.

**Contoh Representative**: NIM 003.

SEGMENT 3: Mahasiswa Kurang Aktif (Low Engagement Students)

**Jumlah**: ±30% dari total mahasiswa

**Ciri Khas** :IPK < 3.0,Kehadiran rendah (kurang dari 75%) ,Partisipasi kelas rendah,Minim akses ke e-learning.

**Contoh Representative**: NIM 002.




**BAGIAN 3: REKOMENDASI STRATEGI**

**SEGMENT 1** – Mahasiswa Unggul (High Achievers)

**Strategi Utama**: Retention & Development

**Tactics:**

- Berikan beasiswa prestasi dan program mentoring

- Libatkan dalam proyek riset atau kompetisi akademik

- Jadikan role model untuk teman sekelas

**Expected Outcome**: Meningkatkan motivasi dan mempertahankan kinerja akademik tinggi.

**SEGMENT 2** – Mahasiswa Konsisten (Moderate Performers)

**Strategi Utama**: Motivation & Performance Boost

**Tactics:**

- Adakan pelatihan manajemen waktu dan belajar efektif

- Dorong partisipasi melalui sistem poin kehadiran

- Tambahkan reward untuk peningkatan IPK semester berikutnya

**Expected Outcome**:
Peningkatan prestasi dan keterlibatan akademik sebesar ≥15%.

**SEGMENT 3** – Mahasiswa Kurang Aktif (Low Engagement Students)

**Strategi Utama**: Re-engagement & Support

**Tactics**:

- Program bimbingan akademik khusus

- Reminder otomatis untuk tugas & kehadiran

- Workshop motivasi dan konseling belajar

**Expected Outcome**:
Peningkatan IPK rata-rata dan kehadiran minimal 80%.
