Studi Kasus 1: Data Mahasiswa Atribut: nama, nim, jurusan Method: tampilkanData() Buat 3 object berbeda.

📘 Penjelasan Singkat

Dalam Java:

Class = cetakan / blueprint dari objek
Object = hasil nyata dari class
Atribut = data yang dimiliki object
Method = fungsi yang dilakukan object

Pada soal:

Atribut:
nama
nim
jurusan
Method:
tampilkanData() → menampilkan data mahasiswa
Buat 3 Object:

Artinya membuat 3 mahasiswa berbeda.

```python
CODING JAVA

public class Mahasiswa {

    // Atribut
    String nama;
    String nim;
    String jurusan;

    // Method untuk menampilkan data
    void tampilkanData() {
        System.out.println("Nama     : " + nama);
        System.out.println("NIM      : " + nim);
        System.out.println("Jurusan  : " + jurusan);
        System.out.println("----------------------");
    }

    public static void main(String[] args) {

        // Object 1
        Mahasiswa mhs1 = new Mahasiswa();
        mhs1.nama = "Andi";
        mhs1.nim = "23001";
        mhs1.jurusan = "Informatika";

        // Object 2
        Mahasiswa mhs2 = new Mahasiswa();
        mhs2.nama = "Siti";
        mhs2.nim = "23002";
        mhs2.jurusan = "Sistem Informasi";

        // Object 3
        Mahasiswa mhs3 = new Mahasiswa();
        mhs3.nama = "Budi";
        mhs3.nim = "23003";
        mhs3.jurusan = "Teknik Komputer";

        // Menampilkan data
        mhs1.tampilkanData();
        mhs2.tampilkanData();
        mhs3.tampilkanData();
    }
}
```

📌 Penjelasan Kode
1. Membuat Class

```python
public class Mahasiswa

Artinya membuat class bernama Mahasiswa.
```

```python
Membuat Atribut

String nama;
String nim;
String jurusan;

Digunakan untuk menyimpan data mahasiswa.
```

**Membuat Method**

void tampilkanData()

Method ini berfungsi menampilkan isi data mahasiswa.

**Membuat Object**
Mahasiswa mhs1 = new Mahasiswa();

Artinya membuat object bernama mhs1.

Begitu juga:

mhs2
mhs3

**Mengisi Data Object**
mhs1.nama = "Andi";

Artinya memberi nilai nama pada object mhs1.

**Menjalankan Method**
mhs1.tampilkanData();

Untuk menampilkan data mahasiswa.

```python
Output Program

Nama     : Andi
NIM      : 23001
Jurusan  : Informatika
----------------------

Nama     : Siti
NIM      : 23002
Jurusan  : Sistem Informasi
----------------------

Nama     : Budi
NIM      : 23003
Jurusan  : Teknik Komputer
----------------------
```

```python
Coding Java Versi Constructor

public class Mahasiswa {

    // Atribut
    String nama;
    String nim;
    String jurusan;

    // Constructor
    Mahasiswa(String nama, String nim, String jurusan) {
        this.nama = nama;
        this.nim = nim;
        this.jurusan = jurusan;
    }

    // Method tampilkan data
    void tampilkanData() {
        System.out.println("Nama     : " + nama);
        System.out.println("NIM      : " + nim);
        System.out.println("Jurusan  : " + jurusan);
        System.out.println("----------------------");
    }

    public static void main(String[] args) {

        // Membuat 3 object dengan constructor
        Mahasiswa mhs1 = new Mahasiswa("Andi", "23001", "Informatika");
        Mahasiswa mhs2 = new Mahasiswa("Siti", "23002", "Sistem Informasi");
        Mahasiswa mhs3 = new Mahasiswa("Budi", "23003", "Teknik Komputer");

        // Menampilkan data
        mhs1.tampilkanData();
        mhs2.tampilkanData();
        mhs3.tampilkanData();
    }
}
```

**Penjelasan Kode**
1. Constructor
Mahasiswa(String nama, String nim, String jurusan)

Artinya saat object dibuat, harus mengisi:

nama
nim
jurusan

2. Keyword this
this.nama = nama;
Artinya:
this.nama = atribut milik class
nama = nilai dari parameter constructor
3. Membuat Object
Mahasiswa mhs1 = new Mahasiswa("Andi", "23001", "Informatika");

Artinya membuat object sekaligus mengisi data.

```python
Tanpa Constructor:

Isi data satu-satu.

mhs1.nama = "Andi";
Dengan Constructor:

Langsung isi saat object dibuat.

new Mahasiswa("Andi", "23001", "Informatika");

➡️ Lebih cepat, rapi, dan sering dipakai.
```
