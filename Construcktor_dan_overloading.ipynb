📘 Apa Itu Overloading?

Overloading dalam Java adalah membuat method atau constructor dengan nama yang sama, tetapi parameter berbeda.

Perbedaannya bisa pada:

jumlah parameter
tipe data parameter
urutan parameter

➡️ Tujuannya supaya nama method tetap sama, tetapi bisa digunakan untuk kebutuhan berbeda.

✅ Contoh Method Overloading

```python
public class Hitung {

    // Method 1
    int tambah(int a, int b) {
        return a + b;
    }

    // Method 2
    int tambah(int a, int b, int c) {
        return a + b + c;
    }

    public static void main(String[] args) {
        Hitung h = new Hitung();

        System.out.println(h.tambah(5, 3));
        System.out.println(h.tambah(5, 3, 2));
    }
}

output :
8
10
```

✅ Contoh Constructor Overloading

```python
public class Mahasiswa {

    String nama;
    int umur;

    // Constructor 1
    Mahasiswa() {
        nama = "Tidak diketahui";
        umur = 0;
    }

    // Constructor 2
    Mahasiswa(String n, int u) {
        nama = n;
        umur = u;
    }
}
```

Constructor sama-sama Mahasiswa(), tapi isi parameter berbeda.
📌 Kenapa Dipakai?

Supaya kode:

lebih rapi
mudah dipahami
fleksibel
nama method tidak perlu banyak
🎯 Kesimpulan

Overloading = nama sama, parameter berbeda.
