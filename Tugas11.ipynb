**PART 1 – Business Context (20%)**

1. Apa tujuan kampanye marketing ini?

Tujuan kampanye adalah untuk membuat pemasaran yang lebih personal, dengan cara menemukan segmentasi customer berbasis perilaku agar Tokopedia dapat:

* Menargetkan pengguna secara lebih akurat

* Meningkatkan conversion rate

* Menyesuaikan promo sesuai kebiasaan user

* Mengurangi pemborosan budget iklan

* Meningkatkan engagement & retention

2. Metode clustering mana yang paling cocok? K-Means atau DBSCAN? Mengapa?

DBSCAN lebih cocok, karena:

* Data user e-commerce biasanya tidak berbentuk cluster bulat seperti asumsi K-Means

* DBSCAN bisa menemukan cluster dengan bentuk tidak beraturan

* DBSCAN otomatis mendeteksi outlier (user dengan perilaku tidak biasa)

* Jumlah cluster tidak perlu ditentukan di awal, berbeda dengan K-Means

* Cocok untuk dataset besar 10.000 user

3. Berapa segment yang Anda harapkan? Mengapa?

3–4 segment utama, karena:

Ecommerce umumnya punya 3 core user groups:

* Light users

* Regular/active users

* High-value/power users

* Segmentasi terlalu banyak membuat kampanye sulit dijalankan

* Segment terlalu sedikit membuat kampanye tidak personal

**PART 2 – Analytical Strategy (30%)**

**Step 1: Preprocessing**
• Variabel yang digunakan

Gabungan demografi + behavior + transaksi:

* Usia

* Lokasi

* Device (Android/iOS)

* Frekuensi app open

* Waktu online

* Click rate

* Jumlah order

* Nilai transaksi

* Kategori favorit

• Perlukah reduksi dimensi?

Ya, perlu.

Mengapa?

* Data memiliki banyak variabel (9 variabel) → sulit divisualisasikan

* PCA membantu mengurangi noise

* Membuat clustering lebih stabil

* Mempercepat proses DBSCAN

**Step 2: Clustering**

* Algorithm: DBSCAN

* Parameters:

1. eps = 0.5 (kedekatan customer)

2. min_samples = 10 (minimal pengguna untuk membentuk cluster)

* Expected segments: 3–4 cluster + 1 outlier group

**Step 3: Visualization**

*Technique: t-SNE (Setelah PCA → 2D)

Purpose:

1. Menampilkan cluster seperti pulau-pulau

2. Mudah melihat outlier

3. Membantu interpretasi segmen marketing

**PART 3 – Segment Interpretation (30%)**

Berdasarkan hasil clustering, diperoleh 3 segmen utama:

**Segment A: Low Engagement Users
Characteristics**

1. Usia muda (18–25)

2. Lokasi acak, mayoritas luar Jawa

3. Frekuensi buka app rendah

4. Nilai transaksi kecil (≤ Rp 100.000/bulan)

* Behavior pattern:

1. Hanya buka app saat promo

2. Jarang checkout

3. Sering klik kategori fashion murah

Size: ± 4.000 users

**Segment B: Regular Shoppers
Characteristics:**

* Usia 25–35

* Pengguna aktif dengan click rate sedang

* Frekuensi order 3–5 kali/bulan

* Transaksi Rp 300.000–800.000/bulan

* Behavior pattern:

* Cenderung belanja kebutuhan rutin (HP accessories, home supplies)

* Responsif terhadap promo gratis ongkir

Size: ± 3.500 users

**Segment C: High-Value Premium Users
Characteristics:**

* Usia 30–45

* Tinggal di kota besar (Jakarta, Bandung, Surabaya)

* Nilai transaksi tinggi (≥ Rp 1.500.000/bulan)

* Device mayoritas iPhone

Behavior pattern:

1. Belanja kategori elektronik & premium goods

2. Lebih peduli kualitas daripada promo

3. Loyal dan sering repeat order

Size: ± 2.200 users
Special Cases

* Outliers ditemukan: Ya

* Kategori:

1. User dengan transaksi sangat tinggi (super spender)

2. User yang login sangat jarang (inactive users)
