💻 Studi Kasus 2: Sistem Perpustakaan (Versi Constructor)
📘 Penjelasan Singkat

Pada studi kasus ini kita membuat class Buku untuk menyimpan data buku di perpustakaan.

Atribut:
judul
pengarang
tahun
stok
Method:
tampilkanInfo() → menampilkan informasi buku

```python
public class Buku {

    // Atribut
    String judul;
    String pengarang;
    int tahun;
    int stok;

    // Constructor
    Buku(String judul, String pengarang, int tahun, int stok) {
        this.judul = judul;
        this.pengarang = pengarang;
        this.tahun = tahun;
        this.stok = stok;
    }

    // Method tampilkan info
    void tampilkanInfo() {
        System.out.println("Judul      : " + judul);
        System.out.println("Pengarang  : " + pengarang);
        System.out.println("Tahun      : " + tahun);
        System.out.println("Stok       : " + stok);
        System.out.println("------------------------");
    }

    public static void main(String[] args) {

        // Object buku
        Buku b1 = new Buku("Laskar Pelangi", "Andrea Hirata", 2005, 10);
        Buku b2 = new Buku("Bumi Manusia", "Pramoedya", 1980, 5);
        Buku b3 = new Buku("Negeri 5 Menara", "Ahmad Fuadi", 2009, 7);

        b1.tampilkanInfo();
        b2.tampilkanInfo();
        b3.tampilkanInfo();
    }
}
```

🖥️ Output

```python
Judul      : Laskar Pelangi
Pengarang  : Andrea Hirata
Tahun      : 2005
Stok       : 10
------------------------

Judul      : Bumi Manusia
Pengarang  : Pramoedya
Tahun      : 1980
Stok       : 5
------------------------

Judul      : Negeri 5 Menara
Pengarang  : Ahmad Fuadi
Tahun      : 2009
Stok       : 7
------------------------
```

📌 Diskusi: Bagaimana Jika Ada 100 Buku?

Jika ada 100 buku, maka tidak efisien jika membuat object satu per satu seperti ini:

```python
Buku b1 = new Buku(...);
Buku b2 = new Buku(...);
Buku b3 = new Buku(...);

Sampai b100.

➡️ Solusinya adalah menggunakan Array atau ArrayList.
```

✅ Contoh Menggunakan Array
Buku[] daftarBuku = new Buku[100];

Artinya kita bisa menyimpan 100 object buku.

```python
Contoh isi
daftarBuku[0] = new Buku("Java", "Andi", 2020, 5);
daftarBuku[1] = new Buku("Python", "Budi", 2021, 3);

Lalu ditampilkan dengan perulangan:
for(int i = 0; i < 100; i++) {
    daftarBuku[i].tampilkanInfo();
}
```

🎯 Kesimpulan

Jika buku hanya sedikit → object biasa cukup.
Jika buku banyak (100 atau lebih) → gunakan:

✅ Array
✅ ArrayList
✅ Database (MySQL)
